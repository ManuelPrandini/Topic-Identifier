```python
'''Method used to read the edge representing as string and return the 
correspettive dictonary with authors id as keys and number of times
that they have used the keywords as value
:edge the string representing a singular edge entry
return a dictionary of the edge'''
def read_edge_ds1(edge):
    map_edge = dict()
    edge = edge[1:len(edge)-1].replace(" ","")
    splitted_edge = edge.split(",")
    for entry in splitted_edge:
        splitted_entry = entry.split(":")
        splitted_entry[0] = splitted_entry[0][1:len(splitted_entry[0])-1]
        map_edge[splitted_entry[0]] = int(splitted_entry[1])
    return map_edge
    
```

```python
'''Method used to load the informations from file ds1(cleaned and divided by year)
such as the nodes(keywords) and the edges(dictionary of authors)
:file the input ds1.tsv file
return set of nodes and list of edges tuples'''
def load_info_ds1(file):
    nodes = set()
    edges = list()
    with open(file,'r',encoding='utf-8') as f:
        for line in f:
            splitted = line.strip().split('\t')
            if(splitted[0] not in nodes):
                nodes.add(splitted[0])
            if(splitted[1] not in nodes):
                nodes.add(splitted[1])
            edges.append((splitted[0], splitted[1], read_edge_ds1(splitted[2]))) 
        f.close()
    return nodes, edges
            
```

```python
'''Method used to load the informations from file ds2(cleaned adn divided by year)
such as the nodes(authors) and the edges(number of collaborations between
two authors)
:file the input ds2.tsv file
return set of nodes and list of edges tuples'''
def load_info_ds2(file):
    nodes = set()
    edges = list()
    with open(file,'r',encoding='utf-8') as f:
        for line in f:
            splitted = line.strip().split('\t')
            if(splitted[0] not in nodes):
                nodes.add(splitted[0])
            if(splitted[1] not in nodes):
                nodes.add(splitted[1])
            edges.append((splitted[0], splitted[1], int(splitted[2])))
        f.close()
    return nodes, edges
```

```python
'''
Method that calculates the number of collaborations of each authors
during a calendar year starting from ds-2.tsv file
:file_ds2 the file where read the lines
return a dictionary that has years as keys and dictionaries of number of authors collaborations as values
'''
def collaborations(file_ds2):
    year_dict_authors = dict()
    #calculate pubblications of each author for a specific year
    with open(file_ds2,'r',encoding='utf-8') as f:
        for line in f:
            splitted = line.strip().split('\t')
            if int(splitted[0]) <= 2018 and int(splitted[0]) >= 2000:
                if splitted[0] not in year_dict_authors.keys():
                    year_dict_authors[splitted[0]] = dict()
                dict_authors = year_dict_authors[splitted[0]]
                dict_authors[splitted[1]] = 1 if splitted[1] not in dict_authors.keys() else dict_authors[splitted[1]] + int(splitted[3])
                dict_authors[splitted[2]] = 1 if splitted[2] not in dict_authors.keys() else dict_authors[splitted[2]] + int(splitted[3])
                year_dict_authors[splitted[0]] = dict_authors
    f.close()
    return year_dict_authors
    

            
```

```python
'''
Method that calculates the number of authors collaborations accumulated up to a certain year
:file_ds2 the file where read lines 
return a dictionary that has years as keys and dictionaries of authors collaborations as values 
'''
def total_collaborations(file_ds2):
    result = dict()
    authors_collaborations = collaborations(file_ds2)
    years = [x for x in sorted(list(authors_collaborations.keys()))]
    result[years[0]] = authors_collaborations[years[0]]
    for y in range(1,len(years)):
        current = authors_collaborations[years[y]]
        previous = result[years[y-1]]
        summed_dict  = {k : current.get(k,0)+previous.get(k,0) for k in current.keys() | previous.keys()}
        result[years[y]] = summed_dict
    return result
        
```

```python
'''
Method that creates the weighted graph of a specific year, starting by list of edges
and the authors relevance dictionary. The weights of the edges are defined as the sum of 
the numbers of time each author has used the keywords in their pubblications multiplied by 
their relevance.
:list_of_edges list of tuple containing the pair of topics and the dictionary of the authors
:authors the author relevance dictionary
return a undirected weighted graph
'''
def create_weighted_graph_ds1(list_of_edges,authors):
    G = net.Graph()
    edges = list()
    for elem in list_of_edges:
        weight = 0
        arg1,arg2,d = elem
        for aut, n in d.items():
            weight+=n*(authors[aut] if aut in authors.keys() else 1)
        G.add_edge(arg1,arg2,weight=weight)
    return G
```

```python
'''
Method that creates a weighted edge of type ds2(authors collaborations graph)
:list_of_edges list of tuple containing the pair of authors and the number of collaborations
return a undirected weighted graph
'''
def create_weighted_graph_ds2(list_of_edges):
    G = net.Graph()
    for elem in list_of_edges:
        arg1,arg2,col = elem
        G.add_edge(arg1,arg2,weight=col)
    return G
```

```python
'''
Method alternative used to creates graphs for each year using
as edges weights the weighted sums of the number of co-occurences
between two keyword in an article of author with the total collaborations
of that author until that year.
:folder_ds1 folder containing ds1 informations files
:folder_ds2 folder containing ds2 informations files
return a dictionary containing graphs for each years
'''
def generate_all_graphs_by_total_collaborations(folder_ds1,folder_ds2):
    result = dict()
    tot_pub = total_collaborations(folder+"ds-2.tsv")
    for year in range(2000,2019):
        #load info from a specific year
        nodes_ds1, edges_ds1 = load_info_ds1(folder_ds1+str(year)+"_ds1.tsv")
        nodes_ds2, edges_ds2 = load_info_ds2(folder_ds2+str(year)+"_ds2.tsv")
        #calculate relevance of authors
        authors = tot_pub[str(year)]
        #create weighted graph
        graph_ds1 = create_weighted_graph_ds1(edges_ds1,authors)
        result[str(year)] = graph_ds1
    return result
```

```python
'''
Method that creates graphs of type ds1 for each year
:folder_ds1 folder containing ds1 informations files
:folder_ds2 folder containing ds2 informations files
return a dictionary containing graphs for each years
'''
def generate_all_graphs_ds1(folder_ds1,authors_centrality):
    result = dict()
    for year in range(2000,2019):
        #load info from a specific year
        nodes_ds1, edges_ds1 = load_info_ds1(folder_ds1+str(year)+"_ds1.tsv")
        #create weighted graph
        W_G = create_weighted_graph_ds1(edges_ds1,authors_centrality[str(year)])
        result[str(year)] = W_G
    return result
```

```python
'''
Method that generate the graphs of type ds2, where the nodes
are the authors and the weights on the edges are the number
of collaborations between them.
:folder_ds2 folder containing ds2 informations files
return a dictionary containing graphs for each years
'''
def generate_all_graphs_ds2(folder_ds2):
    result = dict()
    for year in range(2000,2019):
        nodes_ds2, edges_ds2 = load_info_ds2(folder_ds2+str(year)+"_ds2.tsv")
        graph_ds2 = create_weighted_graph_ds2(edges_ds2)
        result[str(year)] = graph_ds2
    return result
```

```python
'''
Method that normalize the edge weights of each graph
using the minmax formula
:map_graph the dictionary that contains all graphs
'''
def normalize_edges_all_graphs(map_graph):
    for y, G in map_graph.items():
        #find maximum and minimum weight
        edge, m = min(dict(G.edges).items(), key=lambda x: x[1]['weight'])
        minimum = m['weight']
        edge, m = max(dict(G.edges).items(), key=lambda x: x[1]['weight'])
        maximum = m['weight']
        for edge in G.edges(data=True):
            n1,n2,w = edge
            weight = w['weight']
            G.add_edge(n1,n2,weight=(weight-minimum)/(maximum-minimum))
        
```

```python
'''
Method used to select the top k keywords of a graph based on page rank measure
:graph the input graph 
:k number of the top keywords
return a list of top k keywords
'''
def select_top_keywords_pagerank(graph,k):
    result = list()
    page_rank = net.pagerank(graph)
    if k<=len(graph.node):
        for pair in sorted(page_rank.items(),key= lambda x : x[1],reverse=True)[:k]:
            result.append(pair[0])
    else: raise Exception("k is grather than the number of nodes!")
    return result
```

```python
'''
Method used to select the top k keywords of a graph based on hubness measure
:graph the input graph 
:k number of the top keywords
return a list of top k keywords
'''
def select_top_keywords_hubness(graph,k,iterations):
    result = list()
    hubness, authority = net.hits(graph,max_iter=iterations)
    if k<=len(graph.node):
        for pair in sorted(hubness.items(),key= lambda x : x[1],reverse=True)[:k]:
            result.append(pair[0])
    else: raise Exception("k is grather than the number of nodes!")
    return result
```

```python
'''
Method used to select the top k keywords of a graph based on authority measure
:graph the input graph 
:k number of the top keywords
return a list of top k keywords
'''
def select_top_keywords_authority(graph,k,iterations):
    result = list()
    hubness, authority = net.hits(graph,max_iter=iterations)
    if k<=len(graph.node):
        for pair in sorted(authority.items(),key= lambda x : x[1],reverse=True)[:k]:
            result.append(pair[0])
    else: raise Exception("k is grather than the number of nodes!")
    return result
```

```python
'''
Method used to select the top k keywords of a graph based on closeness centrality measure
:graph the input graph 
:k number of the top keywords
return a list of top k keywords
'''
def select_top_keywords_closeness_centrality(graph,k):
    result = list()
    closeness = net.closeness_centrality(graph)
    if k<=len(graph.node):
        for pair in sorted(closeness.items(),key= lambda x : x[1],reverse=True)[:k]:
            result.append(pair[0])
    else: raise Exception("k is grather than the number of nodes!")
    return result
```

```python
'''
Method used to select the top k keywords of a graph based on betweenness centrality measure
:graph the input graph 
:k number of the top keywords
return a list of top k keywords
'''
def select_top_keywords_betweenness_centrality(graph,k):
    result = list()
    betweenness = net.betweenness_centrality(graph,weight='weight')
    if k<=len(graph.node):
        for pair in sorted(betweenness.items(),key= lambda x : x[1],reverse=True)[:k]:
            result.append(pair[0])
    else: raise Exception("k is grather than the number of nodes!")
    return result
```

```python
'''
Method used to select the top k keywords of a graph based on weighted degree centrality measure
:graph the input graph 
:k number of the top keywords
return a list of top k keywords
'''
def select_top_keywords_degree_centrality(graph,k):
    result = list()
    degree = dict(graph.degree(weight='weight'))
    if k<=len(graph.node):
        for pair in sorted(degree.items(),key= lambda x : x[1],reverse=True)[:k]:
            result.append(pair[0])
    else: raise Exception("k is grather than the number of nodes!")
    return result
```

```python
'''
Method that select the top k keywords from each graph using 
a specific modality(eg pagerank)
:map_graph the dict containing each graph for each year
:k the set keywords size
:modality how calculate the k keywords. It can be:
    -pagerank
    -authority
    -hubness
    -betweenness
    -closeness
return a dict with year as key and the correspettive top k keywords
as value
'''
def select_top_keywords_each_graph(map_graph,k,modality="pagerank",iterations=100):
    top_keywords_graph = dict()
    for year, graph in map_graph.items():
        if modality == "pagerank":
            top_keywords_graph[year] = select_top_keywords_pagerank(graph,k)
        elif modality == "authority":
            top_keywords_graph[year] = select_top_keywords_authority(graph,k,iterations)
        elif modality == "hubness":
            top_keywords_graph[year] = select_top_keywords_hubness(graph,k,iterations)
        elif modality == "betweenness":
            top_keywords_graph[year] = select_top_keywords_betweenness_centrality(graph,k)
        elif modality == "closeness":
            top_keywords_graph[year] = select_top_keywords_closeness_centrality(graph,k)
        elif modality == "degree":
            top_keywords_graph[year] = select_top_keywords_degree_centrality(graph,k)
    return top_keywords_graph
        
```

```python
'''
Method that modifies the Spread of Influence Algorithm(Linear Threshold Model). 

For each node not in the active nodes initializes a dictionary of influences coming from sets of the
different k-top keywords and influence it if the sum of the weights of all nodes of that set is 
greater than the threshold. At the end of the algorithm are founded k topics.
:G the input Graph
:S the set of top-k keywords
return a dictionary containing the found topics
'''
def calculate_topics_spreading_algorithm(G,S):
    if(not G.is_directed()):
        G = G.to_directed()
    
    for n in net.nodes(G):
        #assign random threshold to every node
        r = random.random()
        #r = net.degree(G,nbunch=n,weight='weight') / net.degree(G,nbunch=n)
        #print("node "+n+" ",r)
        G.node[n]['threshold'] = r
    
    #topics is the dictionary of infected
    #nodes by each the k-top keywords
    topics = {}
    #add active starting nodes to the dictionary
    #for each keywords
    for n in S:
        topics[n] = set()
        topics[n].add(n)
    active = copy.deepcopy(S)
    #until convergence condition is met
    while True:
        # reset the curently added nodes
        added = []
        
        for n in net.nodes(G):
            #check only inactive nodes
            if n not in active:
                #influence = 0
                #initialize dictionary of influence
                influence = dict()
                for k in topics.keys():
                    influence[k] = 0
                #check all incoming edges
                for edge in G.in_edges(n, data=True):
                    #if the edge comes from an active node
                    for k,s in topics.items():
                        if edge[0] in s:
                            
                            #add edge weight to corrispettive influence
                            influence[k] += float(edge[2]['weight'])
                    
                            #if active weights reach threshold
                            if influence[k] >= G.node[n]['threshold']:
                                #add node to active nodes
                                active.append(n)
                                #add node to the correctly topic
                                topics[k].add(n)
                                #add node to the list of newly
                                #added nodes
                                added.append(n)
            
        #if there are no newly added nodes
        #then we have reached the convergence
        #condition;
        #otherwise trace the added nodes in
        #the current wave
        if len(added) == 0:
            break
    
    
    return topics
```

```python
'''
Method that calculates the Jaccard Similarity between two sets
:set_1 first set of topic's words
:set_2 second set of topic's words
return the similarity
'''
def jaccard_similarity(set_1,set_2):
    intersection = set_1.intersection(set_2)
    union = set_1.union(set_2)
    return float(len(intersection)/(len(union)))
```

```python
'''
Method that calculates the cosine similarity
between two sets
:a first set
:b second set
return the cosine similarity
'''
def cosine_similarity(a,b):
    
    # count word occurrences
    a_vals = Counter(list(a))
    b_vals = Counter(list(b))

    # convert to word-vectors
    words  = list(a_vals.keys() | b_vals.keys())
    a_vect = [a_vals.get(word, 0) for word in words]      
    b_vect = [b_vals.get(word, 0) for word in words]        

    # find cosine
    len_a  = sum(av*av for av in a_vect) ** 0.5             
    len_b  = sum(bv*bv for bv in b_vect) ** 0.5             
    dot    = sum(av*bv for av,bv in zip(a_vect, b_vect))    
    cosine = dot / (len_a * len_b)
    return cosine
```

```python
'''
Method that calculates the overlap similarity
between two sets
:a first set
:b second set
return the overlap similarity
'''
def overlap_similarity(a,b):
    return len(a.intersection(b))/min(len(a),len(b))
```

```python
'''
Method used to merge topics of a specific year through the 
jaccard similarity. The topic similarity must be grather than the threshold similarity.
:topic_year the set of topics of a specific year
:threshold the threshold similarity
return a dictionary with keys and relative sets merged.
'''
def merge_topics_specific_year(topic_year,threshold):
    topic_group = copy.deepcopy(topic_year)

    while True:
        best_similarity = 0
        b_merge = False
        
        for k1,v1 in topic_group.items():
            for k2,v2 in topic_group.items():
                if k1 != k2:
                    #sim = jaccard_similarity(v1,v2)
                    sim = overlap_similarity(v1,v2)
                    #if the calculated similarity is grather than the best_similarity
                    #and the minimal threshold
                    if sim >= best_similarity and sim >=threshold:
                        b_merge = True
                        best_similarity = sim
                        save_k1 = k1
                        save_k2 = k2
                        
        #after comparing all pairs of topics
        #if in the iteration no one pair has been created, exit to the process
        if not b_merge:
            break
        #else, create the merge topic and concatenate the keys
        topic_group[save_k1+"|"+save_k2] = topic_group[save_k1].union(topic_group[save_k2])
        topic_group.pop(save_k1)
        topic_group.pop(save_k2)
    return topic_group
```

```python
'''
Method that generates topics for each year and try to merge them if they 
are similar
:map_graph the dictionary containing graph for each year
:top_keywords the dictionary containing top keywords for each year
return a dictionary containing topics for each year
'''
def generate_all_topics(map_graph,top_keywords):
    topics = dict()
    #generate all separate topics
    for year, graph in map_graph.items():
        topics[year] = calculate_topics_spreading_algorithm(graph,top_keywords[year])
        #try to merge
        topics[year] = merge_topics_specific_year(topics[year],0.55)
        
    return topics
    
```

```python
'''
Method that create pairs of similar topics of consecutive years based
on cosine similarity between us. For each topic of year y find the best 
similar topic of the consecutive year y+1.
:all_topics the dictionary containing topics for each year
'''
def find_similar_topics_consecutive_years(all_topics):
    years = list(all_topics.keys())
    result = list()
    for y in range(len(years)-1):
        for k, v in all_topics[years[y]].items():
            max_sim = 0
            for k1, v1 in all_topics[years[y+1]].items():
                #if the cosine similarity of the two topics are similar
                #sim = cosine_similarity(v,v1)
                sim = overlap_similarity(v,v1)
                if sim >= max_sim:
                    max_sim = sim
                    save_k = k1
                    save_v = v1
            #write the informations about the matching: (only if the cosine similarity is greater than 0)
            #tuple composed by (two consecutive years, the topics keys, first set topic, second set topic, similarity)
            if(max_sim != 0):
                informations = (years[y]+"-"+years[y+1], k, save_k, v , save_v, max_sim)
                result.append(informations)
    return result
                
                    
            
            
```

```python
'''
Method that calculates the similarity between two topics of consecutive
years using the corrispettive subgraphs create with only common nodes, in particular the overlapping similarity
between edges.
:map_subgraphs dictionary containing for each year the keyword of the topic as keys
and the subgraph created with the topic keywords as value
return a list of tuple where a tuple is composed by the two years, the two keys, the two set keywords(topics) 
and the similarity between them
'''
def find_similar_topics_node_intersection_subgraphs(all_topics,map_graph):
    years = list(all_topics.keys())
    result = list()
    for y in range(len(years)-1):
        for k, v in all_topics[years[y]].items():
            max_sim = 0
            for k1, v1 in all_topics[years[y+1]].items():
                #sim = cosine_similarity(v,v1)
                #sim = overlap_similarity(v,v1)
                #takes only nodes in common
                common = v.intersection(v1)
                if len(common) > 1:
                    s_g1 = net.subgraph(map_graph[years[y]],common)
                    s_g2 = net.subgraph(map_graph[years[y+1]],common)
                    edges_g1 = set(s_g1.edges)
                    edges_g2 = set(s_g2.edges)
                    if len(edges_g1) != 0 and len(edges_g2) != 0:
                        sim = overlap_similarity(edges_g1,edges_g2)
                        if sim >= max_sim:
                            max_sim = sim
                            save_k = k1
                            save_v = v1
            if(max_sim != 0):
                informations = (years[y]+"-"+years[y+1], k, save_k, v , save_v, max_sim)
                result.append(informations)
    return result
```

```python
'''
Method that creates for each year and for each topic of a specific year
the subgraph with nodes contained in the topics
:map_graph the dictionary containing all graphs for each year
:all_topics the dictionary containing all topics for each year
return a dictionary that has for each year a dictionary of subgraphs
'''
def create_topics_subgraphs(map_graph,all_topics):
    result = dict()
    years = list(all_topics.keys())
    for y in years:
        subgraphs = dict()
        for k,v in all_topics[y].items():
            subgraphs[k] = map_graph[y].subgraph(list(v))
        result[y] = copy.deepcopy(subgraphs)
    return result
```

```python
'''
Method that produce the merge list of similar topic of consecutive years
produced by the 'find_similar_topics_consecutive_years()' function or
'find_similar_topics_based_on_subgraphs()' function
:similar_topics list of tuples containing years, keys, sets(topics) and the similarity between us
return a list of merged topics
'''
def produce_merge_list_similar_topics_consecutive_year(similar_topics):
    result = list()
    insert_yet = list()
    for el in similar_topics:
        years, key1 , key2 , set_1, set_2, similarity = el
        set_result = set_1.union(set_2)
        if set_result not in insert_yet:
            result.append(set_result)
            insert_yet.append(set_result)
    return result
```

```python
'''
Method that trace all topics during a time interval from a start_date to the
end_date. It groups the topics for the key top keyword, and each top keyword
contains the year and the topic produced in that year.
:all_topics dictionary that contains topics for each year
:start_date where to start to trace topics
:end_date where to finish to trace topics
return a dictionary that has a key the top keyword topic labels, and a dictionary
with year and relative topic as value
'''
def trace_all_topics_time_interval(all_topics,start_date,end_date):
    if start_date > end_date:
        raise(Exception("Error: start date is greater than end date!"))
    result = dict()
    for year in range(start_date,end_date+1):
        for k, t in all_topics[str(year)].items():
            #splitted the key
            split = k.split("|")
            for spl in split:
                if spl not in result:
                    result[spl] = dict()
                result[spl][str(year)] = t
    return result
```

```python
'''
Method that writes the final result in a file.
The file contains the best pairs of topics of consecutive years based
on a specific similarity used. Furthermore, file contains the final merged
list of the merge 'inter-annual' topics.
:filename name of the file
:similar_topics list of tuples that contain two years, the keys, the topic sets, the similarity
:merged_list the list of merge 'inter-annual' topics.
:is_subgraphs boolean that indicates if the input is based on the edge subgraphs similarity
'''
def write_file_final_result(filename,similar_topics,merged_list,similarity_used):
    with open(folder_result+filename,'w') as fw:
        fw.write("PRINT THE SIMILAR TOPICS IN CONSECUTIVE YEARS:"+str.upper(similarity_used)+"\n\n")
        for el in similar_topics:
            years, key1 , key2 , set_1, set_2, similarity = el
            fw.write(years+" --> "+key1+" <--> "+key2+"\n")
            fw.write("1 set: "+str(set_1)+"\n")
            fw.write("2 set: "+str(set_2)+"\n")
            fw.write(similarity_used+": "+str(similarity)+"\n\n")
        
        fw.write("FINAL MERGED LIST:\n\n")
        fw.write(str(merged_list))
            
        fw.close()
        
        
```

```python
'''
Method that calculates the number of times that a keywords appears in the topic set
from the 2000 to 2018.
:all_topics dictionary that contains for each year the relative topics found.
return a dictionary that has a keyword as key and the number of times that appears in the topics.
'''
def keyword_that_appears_more_times(all_topics):
    years = list(all_topics.keys())
    result = dict()
    #for each year
    for y in years:
        #for each topic
        for k, s in all_topics[y].items():
            #for each keyword of topic
            for w in list(s):
                if w not in result:
                    result[w] = 1
                result[w]+=1                
    return result
    
```

```python
'''
Method that calculate the number of times that a topic (based on its label) appears from 2000 to 2018.
:topic_over_year dictionary that contains for each keyword the year which its appears and the topic set
return a dictionary that has keyword as keys and number of appearances as values
'''
def number_topic_appears(topic_over_years):
    res = dict()
    for k, d in topic_over_years.items():
        res[k] = 0
        for y, s in d.items():
            res[k]+=1
    return res
```
